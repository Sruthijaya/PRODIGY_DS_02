```python
import pandas as pd
data=pd.read_csv('titanic.csv')
data.head()
```

DATA CLEANSING

```python
data.isnull().sum()
```

```python
data.drop(columns='Cabin' ,inplace= True)
```

```python
data['Age'].fillna(data['Age'].median() , inplace=True)
```

```python
data['Embarked'].fillna('S',inplace=True)
```

```python
data.isnull().sum()
```

EDA

```python
data.describe()
```

```python
data.info
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6)) 
sns.histplot(data['Age'], bins=10, kde=True, color='coral')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

- The average age of the passengers on the Titanic was likely in the 30s.
- There were more young adults on the Titanic than older adults or children.

```python
#Survival by Passenger Class
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=data, palette='viridis')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

- There were significantly more survivors in first class than in any other class.
- Passengers in first class survived  because they were evacuated first. 

```python
#Survival by Sex
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=data, palette='viridis')
plt.title('Survival Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

There were many more female survivors than male survivors.Specifically, there were more than twice as many female survivors than male survivors.Women and children were evacuated first from the Titanic, which significantly increased their chance of survival.

```python
#Survival by Age
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', hue='Survived',  palette='viridis')
plt.title('Survival by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

The average age of the passengers on the Titanic was likely in the 30s.
There were more young adults on the Titanic than older adults or children.
Children and women were more likely to survive than men.

CONCLUSION

The analysis revealed that the majority of passengers were likely between 20 and 40 years old, with a higher proportion of males on board. 
A clear correlation emerged between travel class and survival rates.
- Passengers in first class had a significantly higher chance of survival compared to those in second and third class. This highlights the impact of social status on evacuation procedures.
- The analysis confirmed that women had a substantially higher survival rate than men.
